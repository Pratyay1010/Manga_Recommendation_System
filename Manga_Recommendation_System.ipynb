```python
def convert2onestring(y):
  return " ".join(list(map(lambda x : x.replace(" ",""), y.strip("'[]'").replace("'","").split(", "))))
```

```python
import ast

def authors_names(names):
  d = pd.DataFrame(ast.literal_eval(names))
  d['name'] = d['first_name'] + d['last_name']

  s = ''
  for i in range(len(d)):
    s += d['name'][i]
    s += " "
  return s
```

```python
new_df['authors'] = new_df['authors'].map(lambda x : authors_names(x))
```

```python
import re

def remove_special_characters(text):
    # Remove symbols and non-alphanumeric characters
    text = re.sub(r'[^\w\s]', '', text)

    # Remove emojis
    text = re.sub(r'[^\w\s,]', '', text)

    # Replace newline characters with spaces
    text = re.sub(r'[\n\r]', ' ', text)

    # Remove extra whitespace
    text = re.sub(r'\s+', ' ', text)

    return text
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()
```

```python
from nltk.stem import SnowballStemmer
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
```

```python
def stem_words(text):
  stemmer = SnowballStemmer('english')

  sentence = text
  words = word_tokenize(sentence)

  stemmed_words = [stemmer.stem(word) for word in words]
  
  return " ".join(stemmed_words)
```

```python
final_df['tags'] = final_df['tags'].map(lambda x : stem_words(x))
```

```python
final_df.head()
```

```python
final_df = final_df.reset_index(drop=True)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(final_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
sorted(list(enumerate(similarity[0])), key=lambda x:x[1], reverse=True)[1:6]
```

```python
def recommend_manga(name):
  index = final_df[final_df['title'] == name].index[0]
  
  similar_mangas = sorted(list(enumerate(similarity[index])), key=lambda x:x[1], reverse=True)[1:6]

  recommendation = []
  for name in similar_mangas:
    recommendation.append(final_df.iloc[name[0]]['title'])

  return recommendation
```

```python
recommend_manga('Berserk')
```
